```python
def show_modules(img, n, m, p1=0, p2=0):
    fig,ax = plt.subplots(n,m,figsize=(12,12))
    for an in range(n):
        for am in range(m):
            im = img[(an+p1)*256:(an+1+p1)*256,(am+p2)*256:(am+1+p2)*256]
            show_image(im, ax=ax[an,am])
```
