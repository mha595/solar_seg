```python
def show_image(img:Image, ax:plt.Axes=None, figsize:tuple=(3,3), hide_axis:bool=True, cmap:str='binary',
                alpha:float=None)->plt.Axes:
    if ax is None: fig,ax = plt.subplots(figsize=figsize)
    ax.imshow(img, cmap=cmap, alpha=alpha)
    if hide_axis: ax.axis('off')
    return ax
```

```python
def show_modules(img, n, m, p1=0, p2=0):
    fig,ax = plt.subplots(n,m,figsize=(12,12))
    for an in range(n):
        for am in range(m):
            im = img[(an+p1)*256:(an+1+p1)*256,(am+p2)*256:(am+1+p2)*256]
            show_image(im, ax=ax[an,am])
```
